# Extra-Mile 1 – Exploring the Installation Dataset

This notebook explores `installed.json`, a dataset containing elevator installation records for devices registered in Ontario. The goals are to understand dataset structure, examine device type distribution, and review device status values.

## (a) Load the dataset and document its structure

We load `installed.json` using `pd.read_json` and inspect its shape, column data types, and the first few rows to understand what information is available.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json("../data/installed.json")

print("Shape:", df.shape)
```

```python
df.dtypes
```

```python
df.head()
```

## (b) Device type distribution

The `Device Type` column identifies what kind of elevator each record describes. We use `value_counts()` to quantify how often each type appears, then visualize the distribution as a horizontal bar chart.

```python
df["Device Type"].value_counts()
```

```python
type_counts = df["Device Type"].value_counts()

plt.figure(figsize=(10, 6))
type_counts.sort_values().plot(kind="barh")
plt.title("Installed Elevator Counts by Device Type")
plt.xlabel("Number of Devices")
plt.ylabel("Device Type")
plt.tight_layout()
plt.show()
```

## (c) Device status distribution

The `DeviceStatus` column records the operational state of each device. We use `value_counts()` to understand how devices are distributed across status values.

```python
df["DeviceStatus"].value_counts()
```

## (d) Comparison with license.csv

Both `installed.json` and `license.csv` contain an elevator identifier column — `Elevating devices number` in installed.json and `ElevatingDevicesNumber` in license.csv — which represents the same device identifier across both datasets and is the basis for any future cross-dataset joins. Both datasets also include a physical location field for each device. However, `installed.json` contains information not present in `license.csv`: the `Device Type` column classifies each elevator (e.g., Passenger Elevator, Freight Elevator, LULA Elevator), and the `Device Class` column groups devices into broader categories such as Elevators. The `DeviceStatus` column in `installed.json` also provides a separate operational status record that is distinct from the license status tracked in `license.csv`. Together, the two datasets complement each other: license.csv captures compliance and licensing data, while installed.json captures physical classification and installation details.
